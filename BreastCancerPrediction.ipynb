```python
from sklearn.neighbors import LocalOutlierFactor

# split the data to X and y before Local Outlier Factorization

y=df["diagnosis"] # Separate the target variable 'diagnosis' from the feature set.
X=df.drop(["diagnosis"],axis=1) # Drop the 'diagnosis' column to isolate the feature set for outlier detection.
columns= df.columns.tolist() # List all column names for reference.
```

```python
lof= LocalOutlierFactor() # Initialize the Local Outlier Factor (LOF) algorithm.
y_pred=lof.fit_predict(X) # Fit the LOF model to the feature set X and predict whether each point is an outlier.
y_pred[0:30] # Display the first 30 predictions where 1 indicates an inlier and -1 indicates an outlier.
#  1 = inlier
# -1 = outlier
```

```python
x_score= lof.negative_outlier_factor_ # Retrieve the outlier factor scores for each sample from the LOF model.
outlier_score= pd.DataFrame()  # Create a DataFrame to store the outlier scores.
outlier_score["score"]=x_score

lofthreshold= -2.5 # Define the threshold for identifying outliers based on the outlier scores.
loffilter= outlier_score["score"]< lofthreshold # Filter the DataFrame to find indices where the outlier score is below the threshold.
outlier_index= outlier_score[loffilter].index.tolist() # Get the indices of samples classified as outliers.
```

```python
plt.figure(figsize=(12,8.)) # Create a figure with a specified size for plotting.
plt.scatter(X.iloc[outlier_index,0],X.iloc[outlier_index,4],color="blue",s=50,label="outliers") # Plot the detected outliers in blue, using specific feature columns for x and y axes.
plt.scatter(X.iloc[:,0],X.iloc[:,4],color="k",s=3,label="Data Points") # Plot all data points in black to visualize the entire dataset.

radius=(x_score.max()- x_score)/(x_score.max()-x_score.min()) # Calculate the radius for each data point based on its outlier score for visual emphasis.
outlier_score["radius"]=radius
plt.scatter(X.iloc[:,0],X.iloc[:,4],s=1000*radius,edgecolors="r",facecolors="none",label="outlier scores") # Plot the outlier scores with varying sizes, using radius for the size of the data points.
plt.legend();
```

```python
X= X.drop(outlier_index) # Remove the rows corresponding to the detected outliers from the feature matrix X.
y= y.drop(outlier_index).values # Remove the labels of the detected outliers from the target vector y and convert it to a NumPy array.
```

```python
# Split data into training and test sets and standardize features.
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NeighborhoodComponentsAnalysis
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)

# Dont fit the scaler while standardizate X_test !
scaler = StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
# Define and initialize a dictionary of different classification models for comparison.
key = ['LogisticRegression','KNeighborsClassifier','SVC','DecisionTreeClassifier','RandomForestClassifier','GradientBoostingClassifier','AdaBoostClassifier','XGBClassifier']
value = [LogisticRegression(), KNeighborsClassifier(n_neighbors = 2, weights ='uniform'), SVC(kernel="rbf",random_state=15), DecisionTreeClassifier(random_state=10), RandomForestClassifier(n_estimators=60, random_state=0), GradientBoostingClassifier(random_state=20), AdaBoostClassifier(), xgb.XGBClassifier(random_state=0,booster="gbtree")]
models = dict(zip(key,value))
models
```

```python
# Train each model on the training data, predict on the test data, and compute the accuracy.
predicted =[]
for name,algo in models.items():
    model=algo
    model.fit(X_train,y_train)
    predict = model.predict(X_test)
    acc = accuracy_score(y_test, predict)
    predicted.append(acc)
    print(name,acc)
```

```python
# Plot the accuracies of different models using a bar chart for comparison.
plt.figure(figsize = (10,5))
sns.barplot(x = predicted, y = key, palette='pastel')
plt.title("Plotting the Model Accuracies", fontsize=16, fontweight="bold")
```

Our analysis shows that all models achieve over 90% accuracy, with the DecisionTreeClassifier at 90.06% and the LogisticRegression reaching 98.25%

```python
lr = LogisticRegression(solver='lbfgs', max_iter=10000) # Initialize the Logistic Regression model and test its performance across different random states.
rs = [] # List to store random state values
acc = [] # List to store accuracy scores
# Iterate through different random states to evaluate model performance
for i in range(1,25,1):
    X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.2, random_state = i)
    model_lr_rs = lr.fit(X_train, y_train)
    predict_values_lr_rs = model_lr_rs.predict(X_test)
    acc.append(accuracy_score(y_test, predict_values_lr_rs))
    rs.append(i)
```

```python
plt.figure(figsize=(10,10)) # Plot accuracy scores against different random states for the Logistic Regression model
plt.plot(rs, acc, color ='red') # Plot accuracy scores with red line

# Print random state values and corresponding accuracy scores
for i in range(len(rs)):
    print(rs[i],acc[i])
```

 The plot reveals that the model achieves accuracy around 99% between the random states 20 and 25. Let's pinpoint the exact values for these states.

```python
for i in range(0,24): # Print accuracy scores that exceed 98% for the Logistic Regression model across different random states
    if acc[i] > 0.98:
        print(acc[i])
```
